```python
# Count how many districts each pincode belongs to in West Bengal
pin_district_count_wb = (
    df_wb
    .groupby('pincode')['district_clean']
    .nunique()
    .reset_index(name='district_count')
)

pin_district_count_wb
```

```python
# Find pincodes mapped to more than one district in West Bengal
problem_pins_wb = pin_district_count_wb[
    pin_district_count_wb['district_count'] > 1
]

problem_pins_wb
```

```python
flagged_wb = df_wb.merge(
    problem_pins_wb[['pincode']],
    on = 'pincode',
    how = 'inner'
)
```

```python
flagged_wb
```

```python
# Group flagged West Bengal data by district & pincode
flagged_pincode_dominant_wb = (
    flagged_wb
    .groupby(['district_clean', 'pincode'])[['age_0_5', 'age_5_17', 'age_18_greater']]
    .sum()
    .reset_index()
)
```

```python
# Calculate total enrolments
flagged_pincode_dominant_wb['total_enroll'] = (
    flagged_pincode_dominant_wb['age_0_5']
    + flagged_pincode_dominant_wb['age_5_17']
    + flagged_pincode_dominant_wb['age_18_greater']
)

# Sort by pincode
flagged_pincode_dominant_wb.sort_values('pincode')
```

```python
# Get dominant district per pincode in West Bengal
idx = (
    flagged_pincode_dominant_wb
    .groupby('pincode')['total_enroll']
    .idxmax()
)

df_filtered_wb = flagged_pincode_dominant_wb.loc[idx]

df_filtered_wb
```

```python
# Avoid SettingWithCopyWarning
df_filtered_wb = df_filtered_wb.copy()

# Flag pincodes mapped to multiple districts in West Bengal
df_filtered_wb.loc[:, 'pin_multi_district_flag'] = (
    df_filtered_wb
    .groupby('pincode')['district_clean']
    .transform('nunique') > 1
)
```

```python
# Ensure df_wb is a proper copy
df_wb = df_wb.copy()

# Extract month from new_date (YYYYMMDD or YYYYMM format)
# df_wb['month'] = df_wb['new_date'].astype(str).str[4:6]
df_wb['month'] = pd.to_datetime(df_wb['new_date']).dt.month

df_wb
```

```python
df_wb = df_wb.copy()

df_wb['total_enroll'] = (
    df_wb['age_0_5']
    + df_wb['age_5_17']
    + df_wb['age_18_greater']
)
```

```python
df_wb_cleaned = df_wb.drop(columns=['date', 'district', 'state'],axis=1)
df_wb_cleaned
```

```python
df_wb = df_wb.copy()

df_wb['total_enroll'] = (
    df_wb['age_0_5']
    + df_wb['age_5_17']
    + df_wb['age_18_greater']
)
```

```python
df_wb_dist_level = (
    df_wb
    .groupby('district_clean')[['age_0_5', 'age_5_17', 'age_18_greater', 'total_enroll']]
    .sum()
)

df_wb_dist_level
```

```python
df_wb_dist_level.sort_values('total_enroll',ascending= False).reset_index()
```
