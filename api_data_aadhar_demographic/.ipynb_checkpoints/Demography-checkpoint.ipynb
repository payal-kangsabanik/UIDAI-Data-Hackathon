```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing Demography 1 File

```python
demo_1=pd.read_csv("api_data_aadhar_demographic_0_500000.csv")
demo_1.head() 
```

```python
demo_1.shape
```

```python
demo_1['state'].unique()
```
